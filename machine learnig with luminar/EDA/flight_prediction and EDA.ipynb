**Concate 2 data**

```python
df_final=pd.concat([df_train,df_test])
df_final.head()
```

```python
df_final.info()
```

**missing value**

```python
df_final.isna().sum()
```

**Handling  Date_of_Journey**

```python
df_final['Date']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final.head(3)
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['year']=df_final['Date_of_Journey'].str.split('/').str[2]
df_final.head(5)
```

**check the data type**

```python
df_final.info()
```

**Date ,month,year object dtype into int**

```python
df_final['Date']=df_final['Date'].astype(int)
df_final['Month']=df_final['Month'].astype(int)
df_final['year']=df_final['year'].astype(int)
df_final.head(1)
```

**Drop the Date of journey row**

```python
df_final.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df_final.head(3)
```

**Handling Arrival Time**

```python
df_final['Arrival_Time']=df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final['Arrival_hour']=df_final['Arrival_Time'].str.split(':').str[0]
df_final.head(4)
```

```python
df_final['Arrival_minutes']=df_final['Arrival_Time'].str.split(':').str[1]
```

**Convert Arrival time ,Arrival hour into int**

```python
df_final['Arrival_hour']=df_final['Arrival_hour'].astype(int)
df_final['Arrival_minutes']=df_final['Arrival_minutes'].astype(int)

```

**Drop Arrival time**

```python
df_final.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df_final.head(9)
```

Handling the Departure Time 

```python
df_final['Dep_hour']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dep_minutes']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dep_hour']=df_final['Dep_hour'].astype(int)
df_final['Dep_minutes']=df_final['Dep_minutes'].astype(int)
df_final.head(3)
```

**Drop Dep_time**

```python
df_final.drop('Dep_Time',axis=1,inplace=True)
```

**Handling Total stop**

```python
df_final['Total_Stops'].unique()
```

```python
df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'nan':1,'4 stops':4})
df_final
```

```python
df_final.drop('Route',axis=1,inplace=True)
df_final.head(3)
```
