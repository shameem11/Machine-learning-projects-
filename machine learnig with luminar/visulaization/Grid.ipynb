```python
import matplotlib.pyplot as plt
```

```python
hight=[140,160,130,149]
weight=[54,75,81,68]
plt.plot(hight,weight)
front1={'family':'serif',"color":'red','size':20}
front2={'family':'serif',"color":'yellow','size':30}
plt.xlabel('hight in cm',fontdict=front1)
plt.ylabel('weight in cm')
plt.title('Studence deatils',fontdict=front2)
plt.grid(axis='y')
```
