```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
#tunnig
from sklearn.model_selection import GridSearchCV

```

```python
clasi=KNeighborsClassifier()
```

```python
para={'n_neighbors':[3,5,7,9,11],'weights':["uniform","distance"]}
clf=GridSearchCV(clasi,para,cv=10,scoring='accuracy')#cv coree validation
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
cls2=KNeighborsClassifier(n_neighbors=9,weights='distance')
cls2.fit(x_train,y_train)
y_pred2=cls2.predict(x_test)
y_pred2
```

```python
print("after tunning accuracey score",accuracy_score(y_test,y_pred2))
```
