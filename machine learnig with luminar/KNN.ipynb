#traing and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
x_train
```

```python
x_test
```

#Normalization

```python
#standard scaler


#min max scaler

#normalized method

#sd Equation   z=(x-u)/s


#u :mean of traning data
#s:standard deviation
#x=data value each row
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)#fit use dymantion ore rangeil
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

impliment mode  




```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train) #model creation
y_pred=model.predict(x_test)#mechine padicho cheak cheyan anu x_test
y_pred
```

```python
y_test
```

```python
pred=model.predict(scaler.transform([[2,100,70,30,0,30.0,0.10,34]]))
print(pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
result=confusion_matrix(y_test,y_pred)
print(result)
score=accuracy_score(y_test,y_pred)
print(score)
```
